```python
import numpy as np
import pandas as pd

cv = lambda x: np.std(x, ddof=1) / np.mean(x) * 100


bd_agua  = pd.read_excel('DB2.xlsx')


df = pd.DataFrame(bd_agua)

bd_agua.describe()

cv(df['PH AGUA'])
```
